### Importing 3d ploting from matplotlib :

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection = '3d')
```

### Ploting 3d lines and points :

```python
import random
ax =plt.axes(projection = '3d')
zline = np.linspace(0,100,1500)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline,yline,zline)
zdata = 15*np.random.random(150) 
xdata = np.sin(zdata)+0.5*np.random.random(150)
ydata = np.cos(zdata)+0.5*np.random.random(150)
ax.scatter3D(xdata,ydata,zdata,c = zdata,cmap = 'Greens');
```

### 3D Countour Plot :

```python
import numpy as np
def f(x,y):
    return np.sin(np.sqrt(x**2 +y**2))
x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.contour3D(X,Y,Z,50,cmap='binary');
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

### Surface Triangulation :

```python
theta = 2*np.pi*np.random.random(1000)
r = 6*np.random.random(1000)
x = np.ravel(r*np.sin(theta))
y = np.ravel(r*np.cos(theta))
z = f(x,y)
ax = plt.axes(projection = '3d')
ax.plot_trisurf(x,y,z,cmap='viridis',edgecolor='none',linewidth=0.5);
```
